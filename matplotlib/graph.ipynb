```python
plt.style.use('ggplot')
x1=[1,4,3]
x2=[2,6,4]
y1=[2,4,6]
y2=[1,3,5]
plt.plot(x1,y1,label="first")
plt.plot(x2,y2,label="second")
plt.legend()
plt.show
```

```python
plt.style.use('ggplot')
x1=[1,4,3]
x2=[2,6,4]
y1=[2,4,6]
y2=[1,3,5]
plt.plot(x1,y1,label="first",linewidth=8)
plt.plot(x2,y2,label="second",linewidth=8)
plt.legend()
plt.show
```
